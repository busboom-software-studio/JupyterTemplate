# Stacking and Unstacking

```
import pandas as pd
import numpy as np
```

```

df = pd.DataFrame({
    'group': [i%3 for i in range(12)],
    'measure': list('ABCD')*3,
    'rand': np.random.uniform(-10,10, 12)
})
df
```

```
us = df.set_index(['group','measure']).unstack()
us
```

```
us.stack()
```
