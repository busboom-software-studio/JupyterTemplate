# Basic Pandas Operations

If you need to get help on any object, you can type the name of hte object with '?' or '??' after it. For instance, execute the next cell ( \<SHIFT\>-Return or the run arrow button ) to get the help for the `head()` function. The help will appear in a new section at the bottom of you screen. The most common operations in Pandas are selecting columns and indexing. Indexing is selecting a subset of a dataframe, such as a single column, a single row, or a collection of rows and columns. There are a lot of ways to index, and some diferent indexing methods look very similar. Indexing Pandas dataframes is based on the way you index a list, but can be a lot more complex. We'll show simple, basic indexing here, and leave the more complex cases for later. 






````
# Getting columns
# You can list all of the columns in a dataframe with the columns property

df.columns
````

Note that the display of the columns has the word 'Index' at the start. This is because the list of coumns is an object that defines an index on the coluns of the dataset. This will be important later. 

````
# Get a single column by indexing with the name of the column

df['stationgroup']
````

````
# Or, use the name of the station as a property
df.stationgroup
````

You can get a subset of the columns by using a list of column names:


````
df[['stationcode','stationgroup','measure_code']]
````

This may look a bit confusing, as it is a list inside an indexer, and both use square brackets. If it helps to understand it is equivalent to:

````
c = ['stationcode','stationgroup','measure_code']
df[c]
````

You can also get rows, although a somewhat different way. Note that when we printed out the columns above ( with `df['stationgroup']` ) , for a single column, there were two columns of numbers displayed. The right side is the value of the elements of the column. The left side is the column index. In this case the index is just the number of the element in the series, but it can be other things, such as dates. Every row has an index, which is the first column displayed when we display the table: 


````
df.head()
````

The bold numbers on the far left of the table, with no column heading, are the index values. The index is used to identify rows. The value of the  cells in the index are known as "labels", and you can reference rows either by the index label, or by the row's integer position, from the stat of the dataframe. You get rows ( and columns ) with an indexer, and there are many indexer. We will focus on two:

* The label indexer, ``.loc``
* The position indexer, ``.iloc``



We can use these to get individual rows, using the ``.loc`` accessor. To get the row with the label '3':

````
df.loc[3]
````

Note that I wrote "the row with the label '3'", not "the third row". The value '3' is the row label, and it is just a coincidence that is also row number 3. It could be something else. 

  The ``.iloc``, or "index location'' indexer, is used to get rows based on their positions in the list. The index for the ``df`` is integers starting from zero, Here is gettin ghte 3rd row, with the ``.iloc`` indexer:


````
df.iloc[3]
````

In this case, ``.loc[3]`` and ``.iloc[3]`` return the same value, but that isn't always true. The labels in the row index don't have to be integers; they can be dates, strings, or numbers that don't correspond to the row position. That's why there are two indexers. One can always work with integer positions, the other one doesn't  

You can also use slice notation on the indexers to get a range of rows, but this is where the two indexers behave diferently. 

The ``.iloc`` indexer behaves just like a list of integers; the indexer will return the values _between_ the values in the slice, so, like the list slice we did above, ``.iloc[3:5]`` will return the values for the rows 3 and 4:



````
df.iloc[3:5]
````

However, the label indexer doesn't work the same way because it does not always work with integers. If it were working with strings, there is no clear way to define what "before" and "after" mean, so the label indexer has to return both the row identified by the starting value and the row identified by the ending value. Thus, unlike ``.iloc``, ``.iloc[3:5]`` will return three rows. 

````
df.loc[3:5]
````

Let's see what is looks like with the index being something else than integers, like the sample date, using the ``set_index()`` function. By default, the set_index method returns a copy of the dataframe, without changing the dataframe. So, you have to assign it to a new value. In this case, we will just re-assign df back to itself.

````
df = df.set_index('sampledate')
````

Now we have a new index on the dataframe. 

````
df.head()
````

With a DateTime index, you can select rows with dates and date ranges


````
df.loc['2005-01-01':'2005-06-01'].head()
````

There is one more really important kind of indexing for selecting a subset of rows, but it requires a little more explaination. One of the most important features of Pandas is that you can operate on a column -- which is formally known as a Series -- as if it were a single value, and the operation will get applied to every element in the series. For instance, if you wanted to multiply the `result` column by 2:


````
df.result * 2
````

The result is a new Series, of the same size as the original `result`, but with every element doubled. 

The same thing works with local operators, such as '<', or '=='. If you write an inequality statement like `df.result > 100`,  you will get a series of True and False values, indicating if the statement was true for each value: 

````
df.result > 100
````

And, you can use that series to index the dataframe, returning all of the rows where the statement is true. So, if you wanted to get all of the rows where ``result`` is greater than 100: 


````
df[df.result > 100]
````
